```python
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import requests
```

```python
df = pd.read_csv("optimism_artical_data.csv")
df.sample(5)
```

```python
artical = []
count = 0
for i in range(len(df['Artical Link'])):

    link = df['Artical Link'].iloc[i]
    html = requests.get(link)
    soup = BeautifulSoup(html.text , "html.parser")
    
    txt = soup.find_all("div" , class_ = "post")
    
    artical.append(txt[0].text.replace("\n" , "").strip() if txt else "N/A")
    count += 1
    print(f"articals scraped ... {count}/{len(df['Artical Link'])}")
```

```python
df['Articals Content'] = artical

data = df.drop(columns= ["Unnamed: 0","Artical desc" , "Artical Views" , "Artical Date" ])
```

```python
print(data['Artical Link'].iloc[34])
print(data["Articals Content"].iloc[34])
```

```python
data.to_csv("optimism_data.csv")
```
